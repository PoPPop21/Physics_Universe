# Decaimiento nuclear ð®
Se modela con la ecuacion diferencial:
$$
\frac{dy}{dt} = -\lambda y \\ y(0) = y_{0}
$$
donde
$$
 y(t) \text{: es el nÃºmero de nÃºcleos.} \\  \lambda > 0  \text{: es la constante de decaimiento.}
$$
La soluciÃ³n analÃ­tica es:
$$
y(t) = y_{0} e^{-\lambda t}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

def simulacion(lam=1.0, y0=1.0, T=5.0, N=50, h=0.8):
  t = np.linspace(0, T, N+1) #arreglo de tiempos desde cero hasta T

  # --- SoluciÃ³n analÃ­tica ð ---
  y_exact = y0 * np.exp(-lam * t)

  # --- Euler explÃ­cito ð ---
  y_exp = np.zeros(N + 1)
  y_exp[0] = y0 # array vacÃ­o
  for n in range(N):
      y_exp[n + 1] = y_exp[n] * (1 - h * lam) # fÃ³rmula iterativa de euler

  # --- Euler implÃ­cito ð ---
  y_imp = np.zeros(N+1)
  y_imp[0] = y0 # otra vez un array vacÃ­o
  for n in range(N):
      y_imp[n+1] = y_imp[n] / (1 + h*lam) # fÃ³rmula iterativa

  # --- GrÃ¡fico ð ---
  plt.figure(figsize=(8,5))
  plt.plot(t, y_exact, 'k-', label='AnalÃ­tica')
  plt.plot(t, y_exp, 'r--', label='Euler ExplÃ­cito')
  plt.plot(t, y_imp, 'b-.', label='Euler ImplÃ­cito')
  plt.xlabel("t")
  plt.ylabel("y(t)")
  plt.title("Decaimiento nuclear")
  plt.legend()
  plt.grid()
  plt.show()

  # --- AnÃ¡lsiis de errores ð ---
  error_exp = np.abs(y_exact - y_exp)# explcÃ­cito
  error_imp = np.abs(y_exact - y_imp) # implicito

  plt.figure(figsize=(8,5))
  plt.plot(t, error_exp, 'r--', label='Error Euler ExplÃ­cito')
  plt.plot(t, error_imp, 'b-.', label='Error Euler ImplÃ­cito')
  plt.xlabel("t")
  plt.ylabel("Error absoluto")
  plt.title("Errores")
  plt.legend()
  plt.grid()
  plt.show()

# --- Widgets ð ---
interact(simulacion,
         lam=widgets.FloatSlider(value=1.0, min=0.1, max=2.5, step=0.1, description='Î»'),
         y0=widgets.FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='y0'),
         T=widgets.FloatSlider(value=5.0, min=1.0, max=10.0, step=0.1, description='T'),
         N=widgets.IntSlider(value=50, min=10, max=100, step=10, description='N'),
         h=widgets.FloatSlider(value=0.8, min=0.0, max=0.8, step=0.01, description='h')
)
```

```python
# --- Error global ð ---
def error_global(lam=1.0, y0=1.0, T=5.0, Nmax=200):
    steps = np.linspace(10, Nmax, 10, dtype=int)  # 10 valores entre 10 y Nmax
    errors_exp, errors_imp, hs = [], [], []

    for N in steps:
        h = T / N
        t_temp = np.linspace(0, T, N+1)
        y_exact_temp = y0 * np.exp(-lam * t_temp)

        # Euler explÃ­cito
        y_e = np.zeros(N+1)
        y_e[0] = y0
        for n in range(N):
            y_e[n+1] = y_e[n] * (1 - h*lam)
        errors_exp.append(np.max(np.abs(y_exact_temp - y_e)))

        # Euler implÃ­cito
        y_i = np.zeros(N+1)
        y_i[0] = y0
        for n in range(N):
            y_i[n+1] = y_i[n] / (1 + h*lam)
        errors_imp.append(np.max(np.abs(y_exact_temp - y_i)))

        hs.append(h)

    # GrÃ¡fico log-log de error global
    plt.figure(figsize=(8,5))
    plt.loglog(hs, errors_exp, 'ro-', label='Euler ExplÃ­cito')
    plt.loglog(hs, errors_imp, 'bo-', label='Euler ImplÃ­cito')
    plt.xlabel("TamaÃ±o de paso h")
    plt.ylabel("Error global (norma infinito)")
    plt.title("Error global vs tamaÃ±o de paso")
    plt.legend()
    plt.grid(True, which="both")
    plt.show()


# Widgets interactivos
interact(
    error_global,
    lam=widgets.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description="Î»"),
    y0=widgets.FloatSlider(value=1.0, min=0.5, max=5.0, step=0.5, description="yâ"),
    T=widgets.FloatSlider(value=5.0, min=1.0, max=10.0, step=0.5, description="Tiempo T"),
    Nmax=widgets.IntSlider(value=200, min=50, max=1000, step=50, description="N mÃ¡ximo")
)
```
